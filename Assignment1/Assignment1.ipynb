```python
from typing import List, Dict
import re
from collections import defaultdict

class BPETokenizer:
    """Byte Pair Encoding Tokenizer

    This tokenizer implements the Byte Pair Encoding (BPE) algorithm for tokenization

    Args:
        vocab_size (int): the desired size of the vocabulary

    Attributes:
        vocab_size (int): the desired size of the vocabulary
        vocab (Dict[str, int]): the vocabulary mapping tokens to their frequencies
    """

    def __init__(self, vocab_size: int) -> None:
        """initialize the BPETokenizer with a given vocabulary size"""
        self.vocab_size = vocab_size
        self.vocab: Dict[str, int] = {}

    def _calculate_stats(self, vocab: Dict[str, int]) -> Dict[tuple, int]:
        """calculate statistics of token pairs within the vocabulary"""
        pairs = defaultdict(int)
        for word, freq in vocab.items():
            symbols = word.split()
            for i in range(len(symbols) - 1):
                pairs[symbols[i], symbols[i + 1]] += freq
        return pairs

    def _merge_tokens(self, pair: tuple, v_in: Dict[str, int]) -> Dict[str, int]:
        """merge tokens based on the given pair and vocabulary"""
        v_out = {}
        bigram = re.escape(' '.join(pair))
        pattern = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        for word in v_in:
            updated_word = pattern.sub(''.join(pair), word)
            v_out[updated_word] = v_in[word]
        return v_out

    def train(self, corpus: List[str]) -> None:
        """train the tokenizer using the provided corpus"""
        vocab_freq = defaultdict(int)
        for word in corpus:
            symbols = word.split()
            for symbol in symbols:
                vocab_freq[' '.join(symbol)] += 1
                
        # merge tokens until reaching the specified vocabulary size
        for _ in range(self.vocab_size):
            pairs = self._calculate_stats(vocab_freq)
            if not pairs:
                break
            vocab_freq = self._merge_tokens(max(pairs, key=pairs.get), vocab_freq)
        self.vocab = vocab_freq
        
    def tokenize(self, text: str) -> List[str]:
        """tokenize the given text using the trained vocabulary."""
        tokens = []
        for word in text.split():
            symbols = word.split()
            for symbol in symbols:
                tokens.append(symbol) if symbol in self.vocab else tokens.extend(list(symbol))
        return tokens
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000]. Plot the dependency of the average length of the tokenized tweet by `vocab_size` to analyze how `vocab_size` affects the length of the tokenized tweet on average. Tell what `vocab_size` is preferrable and why.

```python
import matplotlib.pyplot as plt

# define the vocabulary sizes to be tested
vocab_sizes = [100, 250, 500, 750, 1000, 2000, 3000, 5000]
avg_token_lengths = []

for vocab_size in vocab_sizes:
    print(f'vocabulary size {vocab_size}')
    tokenizer = BPETokenizer(vocab_size=vocab_size)
    tokenizer.train(merged_sentences)
    tokens = tokenizer.tokenize(' '.join(merged_sentences))
    token_lengths = [len(tokenizer.tokenize(tweet)) for tweet in merged_sentences]
    avg_token_length = sum(token_lengths) / len(token_lengths)
    avg_token_lengths.append(avg_token_length)
    print("tokens:", tokens[:100])

# plotting results
plt.plot(vocab_sizes, avg_token_lengths, marker='o')
plt.xlabel('Vocabulary Size')
plt.ylabel('Average Token Length')
plt.title('Dependency of Average Token Length on Vocabulary Size')
plt.grid(True)
plt.show()
```
